# drawing Probabilty mass function

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import random 
```

```python
L = []
for i in range(1000):
    L.append(random.randint(1,6))
```

```python
len(L)
```

```python
L[:10:]
```

```python
# finding the probabilty of all the outcomes
prob = pd.Series(L).value_counts(normalize=True) # Note : Here noramlize convert the value_counts value to true.
```

```python
prob
```

```python
# doubt -> Why 1 has maximum probability
x = ['1','2','3','4','5','6']
plt.bar(x,prob)
```

```python
# Draw the bar graph of probability if we are trowing two dices simultaneously
L2 = []
for i in range(1000):
    a = random.randint(1,6)
    b = random.randint(1,6)
    L2.append(a+b)

    
```

```python
L2[:10]
```

```python
prob2 = pd.Series(L2).value_counts(normalize=True)
# Reason why different probabilities are coming 
# More combinations → Higher probability
# Fewer combinations → Lower probability
```

```python
prob2
```

```python
x = ['7','8','5','6','9','10','4','11','3','12','2']
plt.bar(x,prob2)
```

# Draw the graph of the cummulative distribution function


```python
import numpy as np
```

```python
from numpy.random import normal  # This function is used for creating random value 
```

```python
# loc == mean , scale = standard_deviation , size == number of element generation 
sample = normal(loc=50,scale=5,size=1000)
```

```python
sample.size
```

```python
# sample[starting_point,ending_point]
sample[0:11]
```

```python
sample.mean()
```

```python
# x-axis is represets the generates values and y axix is represing the frequency of that value
plt.hist(sample,bins=10)
```

```python
sample_mean = sample.mean()
sample_std = sample.std()
```

```python
sample_mean
```

```python
sample_std
```

```python
# fit the distribution with in normal distribution equation
from scipy.stats import norm
dist = norm(sample_mean,sample_std)
```

```python
sample.min()
```

```python
sample.max()
```

```python
# it is used for generating random number between the min and the max provided
value  = np.linspace(sample.min(),sample.max(),100)
```

```python
prob_density = [dist.pdf(value) for value in value]
```

```python
# ploting the graph for histogram and pdf 
plt.hist(sample,bins=10,density=True)
plt.plot(value,prob_density)
```
